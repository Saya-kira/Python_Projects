```python
def cipher_function(Ri,Subkeyi):

    E = [32,1,2,3,4,5,
          4,5,6,7,8,9,
          8,9,10,11,12,13,
         12,13,14,15,16,17,
         16,17,18,19,20,21,
         20,21,22,23,24,25,
         24,25,26,27,28,29,
         28,29,30,31,32,1]
    
    S_1 = [[14,4,13,1,2,15,11,8,3,10,6,12,5,9,0,7],
           [0,15,7,4,14,2,13,1,10,6,12,11,9,5,3,8],
           [4,1,14,8,13,6,2,11,15,12,9,7,3,10,5,0],
           [15,12,8,2,4,9,1,7,5,11,3,14,10,0,6,13]]

    S_2 = [[15,1,8,14,6,11,3,4,9,7,2,13,12,0,5,10],
           [3,13,4,7,15,2,8,14,12,0,1,10,6,9,11,5],
           [0,14,7,11,10,4,13,1,5,8,12,6,9,3,2,15],
           [13,8,10,1,3,15,4,2,11,6,7,12,0,5,14,9]]

    S_3 = [[10,0,9,14,6,3,15,5,1,13,12,7,11,4,2,8],
           [13,7,0,9,3,4,6,10,2,8,5,14,12,11,15,1],
           [13,6,4,9,8,15,3,0,11,1,2,12,5,10,14,7],
           [1,10,13,0,6,9,8,7,4,15,14,3,11,5,2,12]]

    S_4 = [[7,13,14,3,0,6,9,10,1,2,8,5,11,12,4,15],
           [13,8,11,5,6,15,0,3,4,7,2,12,1,10,14,9],
           [10,6,9,0,12,11,7,13,15,1,3,14,5,2,8,4],
           [3,15,0,6,10,1,13,8,9,4,5,11,12,7,2,14]]

    S_5 = [[2,12,4,1,7,10,11,6,8,5,3,15,13,0,14,9],
           [14,11,2,12,4,7,13,1,5,0,15,10,3,9,8,6],
           [4,2,1,11,10,13,7,8,15,9,12,5,6,3,0,14],
           [11,8,12,7,1,14,2,13,6,15,0,9,10,4,5,3]]

    S_6 = [[12,1,10,15,9,2,6,8,0,13,3,4,14,7,5,11],
           [10,15,4,2,7,12,9,5,6,1,13,14,0,11,3,8],
           [9,14,15,5,2,8,12,3,7,0,4,10,1,13,11,6],
           [4,3,2,12,9,5,15,10,11,14,1,7,6,0,8,13]]

    S_7 = [[4,11,2,14,15,0,8,13,3,12,9,7,5,10,6,1],
           [13,0,11,7,4,9,1,10,14,3,5,12,2,15,8,6],
           [1,4,11,13,12,3,7,14,10,15,6,8,0,5,9,2],
           [6,11,13,8,1,4,10,7,9,5,0,15,14,2,3,12]]

    S_8 = [[13,2,8,4,6,15,11,1,10,9,3,14,5,0,12,7],
           [1,15,13,8,10,3,7,4,12,5,6,11,0,14,9,2],
           [7,11,4,1,9,12,14,2,0,6,10,13,15,3,5,8],
           [2,1,14,7,4,10,8,13,15,12,9,0,3,5,6,11]]
    
    collection_of_S = [S_1,S_2,S_3,S_4,S_5,S_6,S_7,S_8]
    
    P = [16,7,20,21,
         29,12,28,17,
          1,15,23,26,
          5,18,31,10,
          2,8,24,14,
         32,27,3,9,
         19,13,30,6,
         22,11,4,25]
    
    ##STEP 1 : APPLYING THE "E" PERMUTATION ON Ri. 
    ##This whill extend the original 32 bits of Ri into a 48 bits called E_Ri. 
    
    E_Ri = []
    
    for i in range(48):
        E_Ri.append(Ri[E[i]-1])
    

    ##STEP 2 : COMPUTING E_Ri XOR Subkeyi :
    XORed = []
    
    for i in range(48) :
        XORed.append(E_Ri[i]^Subkeyi[i])

    ##STEP 3 : CUTTING THE XORed i INTO 8 BLOCKS OF 6bits EACH.
    collection_of_6bits = [] #I'll be putting all those 8 blocks in this list.
                             #It's a list of lists.
    
    for i in range(8):
        collection_of_6bits.append(XORed[i*6:i*6+6])
    
    ##STEP 4 : applying the S_boxes to each blocks.
    row = 0
    column = 0
    collection_of_4bits = [[],[],[],[],[],[],[],[]]
    final_message = []
    
    for i in range(8): #this following operation will be done for each block
        
        ##SUBSTEP 1 : EXTRACTING THE ROW AND COLUMNS
        
        #First we stick x1 and x6 together into a string
        row_string = str(collection_of_6bits[i][0])+str(collection_of_6bits[i][5])
        
        #Then we transform that string of "binary number" into an integer in base 10
        row = int(row_string, 2)
    
        #Same for the column; we need to stick x2x3x4x5 into a string then transform
        #that string into an integer in base 10 to give us the column number
        column_string = str(collection_of_6bits[i][1])+str(collection_of_6bits[i][2])+str(collection_of_6bits[i][3])+str(collection_of_6bits[i][4])
        column = int(column_string, 2)
        
       # value_in_S =  #Note : here I-m collecting the value from the S table
                                                     #I still have to turn it into binary. See next line.
        value_in_S = list(map(int,list(format(collection_of_S[i][row][column], '0>4b')))) 
        
        
        ##SUBSTEP 2 : APPLYING THE FINAL PERMUTATION P
        
        
        for j in range(4):

            final_message.append(value_in_S[j])

    
    
    ##STEP 5 : APPLYING THE FINAL PERMUTATION "P" ON final_message.
    ##I save the last result in :
    Ri_out = []
    for i in range(32):
        Ri_out.append(final_message[P[i]-1])
    return Ri_out
```
