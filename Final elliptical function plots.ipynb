```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import special

u = np.linspace(0.1,10)
k = 0.999
y = sp.special.ellipj(u,k)

plt.plot(u, y[0], label = "sn(u,0.1)")
plt.plot(u,y[1], label = "cn(u,0.1)")
plt.plot(u,y[2], label = "dn(u,0.1)")
plt.xlabel("u")
plt.legend()
plt.grid()
plt.show()
```
