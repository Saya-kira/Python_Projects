```python
#############READ ME################
#
#
#
#
#
## This is what happends in one round:
## you can play many rounds to stack up your winnings.
## if you are satisfied you can end the game.
## See below how to do this.
#
#
#
#
#
## There are 2 players: Human and computer dealer.
## We are using a deck of 52 cards.
#
#
#
#
## The human has a bank roll. (find out what it is and the explain it here.)
## He places a bet, indicating wether he will win this set of hand.
#
#
#
#
#
## The dealer deals 2 cards: 1 card face-up and the other face-down
## The human deals 2 cards: both face-up
#
#
#
#
#
## GOAL: The sum of the values of the two cards that the player placed has to be closest to 21.
## There are 2 ways to win.
#
#
#
#
#
## Human always decides his move first. He can count the sum of both his cards.
## Option 1: He chooses to add another card from the deck. (HIT)
## Option 2: He does nothing. 
## If he does nothing, the dealer reveals his second card 
## and they compare who is closest to 21.
## The winner is then allowed to collect the money in the betting.
#
#
#
#
#
## There are 3 ways the game can end:
## Case 1: the human busts, meaning he has no more money to make a bet.
## (to do this, at the end of each round, the computer will check that the human
## has enough money to place a minimum bet for the next round).
## Computer collects money.
## Case 2: User decides to leave the game.
## (to do this, at the end of each round, after the computer checking that the human
## has enough money, the computer will ask the question if the human wishes to continue)
#
#
#
#
#
## Note: This version has no adjustments for Aces.
###########################################################################################################################
```

```python
# CARD
# SUIT,RANK,VALUE

#These are the global variables.

import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
game_on = True




#This part is complete and correct

class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
    
#############################################################################################################    
#This part is complete and correct

class Deck():
    
    def __init__(self):
        
        self.all_cards = []
        
        for i in suits:
            for j in ranks:
                created_card = Card(i,j)
                self.all_cards.append(created_card)
                
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self): #If we call this method on our deck, 
                        #it will remove a the card from the deck completely.
        return self.all_cards.pop()

#############################################################################################################
    
#STEP 4: This part is incomplete but correct

class Hand(): #These are the cards that have been dealt on the table
    
    def __init__(self):
        self.cards = []
        self.value = 0 #this is the sum of our hand.
        #self.aces = 0 #attribution to keep track of aces.
        
    def add_card(self,card):
        self.cards.append(card)
        self.value = self.value + card.value
    
    #def adjust_for_ace(self):
    #    if  ???

    def reveal(self,position):
        print(self.cards[position])

    
    def __str__(self):
        return f"The sum of the hand is {self.value}." 
    

#############################################################################################################
#STEP 5: This part is correct
#NOTE: This part needs to be used with STEP 6.
 

class Chips:
    
    def __init__(self,initial_pocket_cash):
        self.total = initial_pocket_cash #This is how much the player has initially.
        self.bet = 0 #this will be modified with STEP 6.
        
    def win_bet(self):
        self.total = self.total + self.bet
    
    def lose_bet(self):
        self.total = self.total - self.bet
        
#############################################################################################################
#STEP 6: THIS PART ISN'T FUNCTIONING
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet? '))
        except ValueError:
            print('Sorry, a bet must be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)
            else:
                break


#############################################################################################################
#STEP 7: This part is complete and correct
def hit(deck,hand):
    
    card_hit = deck.deal_one()
    print("\nThe dealt card is the",card_hit)
    hand.add_card(card_hit)
    print(hand)
    
    return
#############################################################################################################
#STEP 8: This part asks the player to decide wether he wants to hit or stand.

def hit_or_stand(deck,player_hand,asking):
    
    decision = input(f"Your current sum is {player_hand.value}. Do you wish to hit? Type Yes or No: ")
        
    if decision=="Yes":
        hit(deck,player_hand)

    if decision=="No":
        asking=False

        
    return asking
    
```
