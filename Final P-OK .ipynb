```python
def prime_generator(): ##FYI : I'm recycling this from my previous TP8 RSA
    import random, sympy, math, numpy

        ##NOTES : I found this module on https://www.geeksforgeeks.org/prime-functions-python-sympy/
        ##It allows me to automatically generate a random prime in a range [a,b].
    
    p = sympy.randprime(10**5,10**10) #Generating this number takes a while.
    q = sympy.randprime(10**5,10**10) #This is our second prime but we need to make sure its premier to q

    ##This part just makes sure that the previous primes are distincts.
    while math.gcd(p,q)!=1:
        q = sympy.randprime(10**5,10**10)
        
    return p*q

def generating_A_secrets():
    
    import random
    k=random.randint(1,100) #k is the number of secrets we will have.
    A_secrets = []
    
    ##Let's generate those k random integers for our secrets.
    ##No choice. We have to use the random.randint function.
    
    for i in range(k):
        A_secrets.append(random.randint(2**5,2**10))
    
    return A_secrets,k

def generating_A_public(secrets,n):
    
    A_public = []
    
    for i in secrets:
        A_public.append((i**2)%(n))
    
    return A_public

    #######################################################################################################
    #######################################################################################################
    #######################################################################################################

    
n = prime_generator()
(A_secrets,k) = generating_A_secrets()
A_public = generating_A_public(A_secrets,n)

def POK(n,k,A_secrets,A_public):
    ##Let's do the POK step-by-step as it is given in the TP11:
    
    def A_generates_r(n):
        import random
    ##STEP 1 : A CHOOSES A RANDOM r, COMPUTES AND SENDS x
        r = random.randint(1,2**10)
        x = (r**2)%n
        return r,x
    
    #######################################################################################################
    
    ##STEP 2 : B SENDS A LIST OF CHALLENGES.
    def generating_B_challenges(k):
        import random
        B_challenges = []
    
        for i in range(k):
            B_challenges.append(random.randint(0,1))
            
        return B_challenges

    #######################################################################################################
    
    ##STEP 3 : A RESPONDS BY COMPUTING y.
    
    def generating_A_y(A_secrets,B_challenges,k,r,n):
 
        temp = r
        for i in range(k):
            temp = temp*((A_secrets[i]**B_challenges[i]))

        y = temp%n

        return y
    
    
    #######################################################################################################

    ##This is where we do all the steps now.
    (x,r) = A_generates_r(n)
    B_challenges = generating_B_challenges(k)
    A_y = generating_A_y(A_secrets,B_challenges,k,r,n)
    
    
    ##So now it's up to B to verify A's identity :
    
    ##First, we need to compute A's y.
    
    y = r
    for i in range(k):
        y = y*(A_secrets[i]**B_challenges[i])%n
    
    ##Then we need to compute the y squared that will be used by B to compare both results.
    
    B_y_squared = x
    for i in range(k):
        B_y_squared = B_y_squared*(A_public[i]**B_challenges[i])%n
    
    
    if (y!=0) and (y**2==B_y_squared):
        print("Identity approved by B")
    else :
        print("Identity rejected by B")
    

POK(n,k,A_secrets,A_public)

```
